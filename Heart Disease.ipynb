# IMPORTING DATASET

```python
df_heart=pd.read_csv('heartdisease_data.csv')
df_heart
```

IMPUTER

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp = imp.fit(df_heart.values)
df=imp.transform(df_heart)
df_heart=pd.DataFrame(df)
df_heart
```

```python
num = df_heart._get_numeric_data()
num[num < 0] = 0
df_heart
```

CHECKING DATA SKEWNESS

```python
for col in df_heart.columns:
    if df_heart.skew().loc[col]>0.55:
        df_heart[col]=np.log1p(df_heart[col])
df_heart.skew()
```

CHECKING OUTLIERS

```python
from scipy.stats import zscore
z=np.abs(zscore(df_heart))
print(z.shape)
print(np.where(z>3))
```

REMOVING OUTLIERS

```python
df_final=df_heart.loc[(z<3).all(axis=1)]
print(df_final.shape)
```

# TRAINING AND TESTING - VARIOUS MODELS

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10,stratify=y)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

CLASSIFIERS

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
knn=KNeighborsClassifier(n_neighbors=12)
sv=SVC()
lr=LogisticRegression()
dtc=DecisionTreeClassifier(random_state=10)
mnb=MultinomialNB()
rfc=RandomForestClassifier()
```

```python
models=[]
models.append(('KNeighborsClassifier',knn))
models.append(('SVC',sv))
models.append(('DecisionTreeClassifier',dtc))
models.append(('LogisticRegression',lr))
models.append(('MultinomialNB',mnb))
models.append(('RandomForestClassifier',rfc))
```

FUNCTION DECLARATION

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
for name,model in models:
    print('*************',name,'*************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    y_pred=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,y_pred)
    print('Accuracy Score = ',AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='accuracy').mean()
    print('cross_val_score = ',sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,y_pred))
    print('\n')
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,50))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    print('\n\n')
```

# BEST MODEL

```python
result=pd.DataFrame({'Model ':Model,'Accuracy Score ':score,'Cross_Val_Score':cvs})
result
```

I choose Logistic Regression for this dataset.
