# IMPORTING DATASET

```python
df_food=pd.read_excel('Data_Train.xlsx')
df_food
```

IMPUTER

```python
imp=SimpleImputer(strategy="most_frequent")
df_food['RATING']=imp.fit_transform(df_food['RATING'].values.reshape(-1,1))
df_food
```

```python
df_food['RATING']=df_food.RATING.str.replace("NEW","0")
df_food[['RATING']] = df_food[['RATING']].fillna(value='0')
df_food[['VOTES']] = df_food[['VOTES']].fillna(value='0')
df_food['VOTES']=df_food.VOTES.str.replace(" votes","")
df_food
```

```python
def doc(series):
    Cuisine = ''
    for i in series:
        Cuisine += i + ','
    return Cuisine
text=doc(df_food.CUISINES)
df=pd.DataFrame.from_dict(dict(Counter(text.split(',')).most_common()), orient='index').reset_index()
df.columns=['CUISINES','Count']
df.head(10)
```

```python
text=doc(df_food.CUISINES)
df=pd.DataFrame.from_dict(dict(Counter(text.split(',')).most_common()), orient='index').reset_index()
df.columns=['CUISINES','Count']
df['Code'] = df.CUISINES.astype('category').cat.codes
df.head(10)
```

```python
cuisines_dict = dict(zip(df.CUISINES, df.Code))
```

```python
dat_train = df_food.CUISINES.apply(lambda x: len(x.split(',')))
def cuisines_col(dataframe, col, col_num):
    return dataframe[col].str.split(',').str[col_num]
for i in range(0,dat_train.max()):
    cuisine = "Cuisine_"+ str(i+1)
    df_food[cuisine] = cuisines_col(df_food,'CUISINES', i)
df_food.head()
```

```python
cols=list(df_food.columns.values)
cols.pop(cols.index('COST'))
df_food=df_food[cols+['COST']]
df_food
```

```python
df_food.drop(df_food.columns[[1,2,3]],axis=1,inplace=True)
```

LABEL ENCODING

```python
le=LabelEncoder()
list=['TITLE','CITY','LOCALITY']
for val in list:
    df_food[val]=le.fit_transform(df_food[val].astype(str))
df_food
```

```python
column_food=['Cuisine_1','Cuisine_2','Cuisine_3','Cuisine_4','Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']
for i in column_food:
    df_food.replace({i: cuisines_dict}, inplace=True)
df_food
```

```python
df_food[['Cuisine_1','Cuisine_2','Cuisine_3','Cuisine_4','Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']] = df_food[['Cuisine_1','Cuisine_2','Cuisine_3','Cuisine_4','Cuisine_5','Cuisine_6','Cuisine_7','Cuisine_8']].fillna(value='0')
df_food
```

# TRAINING AND TESTING - VARIOUS MODELS

FUNCTION DECLARATION

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("R2 Score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("Maximum R2 Score corresponds to ",final_r_state," is ",max_r_score)
    return(final_r_state)
```

LOGISTIC REGRESSION

```python
lr=LinearRegression()
r_state=maxr2_score(lr,x,y)
```

```python
print("Mean R2 Score for Linear Regression is ",cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2').mean())
print("Standard Deviation in R2 Score for Linear Regression is ",cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2').std())
```

```python
knr=KNeighborsRegressor(n_neighbors=8)
r_state=maxr2_score(knr,x,y)
```

```python
print("Mean R2 Score for KNN Regression is ",cross_val_score(knr,x,y,cv=5,scoring="r2").mean())
print("Standard Deviation in R2 Score for KNN Regression is ",cross_val_score(knr,x,y,cv=5,scoring='r2').std())
```

GRADIENT BOOSTING REGRESSOR

```python
gbr=GradientBoostingRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(gbr,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

# BEST MODEL

R2 SCORE AND RMSE

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=61,test_size=0.20)
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
y_pred=gbr.predict(x_test)
print("R2 Score is ",r2_score(y_test,y_pred))
print("RMSE is ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

CROSS VALIDATION

```python
score=cross_val_score(gbr,x,y,cv=5)
print('Score is ',score)
print('Mean Score is ', score.mean())
print('Standard Deviation is ',score.std())
```
